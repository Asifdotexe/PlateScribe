```python
import cv2 as cv
import plotly.express as px
from skimage import io
import matplotlib.pyplot as plt
import xml.etree.ElementTree as xet
%matplotlib inline
```

```python
# reading the image using skimage
image = io.imread(r'..\input\number-plates\images\Cars0.png')
```

```python
# finding cordinate values
xmin = int(xet.parse(r'..\input\number-plates\annotations\Cars0.xml')\
    .getroot()\
        .find('object')\
            .find('bndbox')\
                .find('xmin').text)

xmax = int(xet.parse(r'..\input\number-plates\annotations\Cars0.xml')\
    .getroot()\
        .find('object')\
            .find('bndbox')\
                .find('xmax').text)

ymin = int(xet.parse(r'..\input\number-plates\annotations\Cars0.xml')\
    .getroot()\
        .find('object')\
            .find('bndbox')\
                .find('ymin').text)

ymax = int(xet.parse(r'..\input\number-plates\annotations\Cars0.xml')\
    .getroot()\
        .find('object')\
            .find('bndbox')\
                .find('ymax').text)
```

```python
# storing the image
fig = px.imshow(image)
```

```python
fig.update_layout(
    width=600, height=500, margin=dict(l=10, r=10, b=10, t=10),
    xaxis_title='Bounding Box Verification')
```

```python
fig.add_shape(
    type="rect", xref="x", yref="y",
    x0=xmin, y0=ymin, x1=xmax, y1=ymax,
    line_color="cyan"
)
```
